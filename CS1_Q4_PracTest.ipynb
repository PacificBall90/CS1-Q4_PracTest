```python
inputs_done = False
scores = []
items = []
counter = 0

def average(sum, count):
  return sum / count

while inputs_done == False:
 counter = counter + 1
 scores.append(int(input("Please enter an assessment score: ")))
 items.append(int(input("And how many items were there?: ")))
 ans = input("Are there any more scores you would like to input? (Y, N): ")
 if ans == "Y" or ans == "y":
  pass
 else:
  inputs_done = True
 print()

print("Calculating...")
print()

total_items = 0
for x in items:
  total_items = total_items + x

total_score = 0
for x in scores:
  total_score = total_score + x

print("Your average score is:", average(total_score, counter))
print("That is", (average(total_score, counter) / average(total_items, counter)) * 100,"% of the average items")
print()

if average(total_score, counter) / average(total_items, counter) >= .6:
  print("You passed!")
else:
  print("You failed...")
```
